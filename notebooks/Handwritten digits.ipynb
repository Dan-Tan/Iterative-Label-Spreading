# Clustering on hand writtern digits

This notebook seeks to go over the steps one might take to cluster a data set.

### Load Data and shape into numpy object with (n_samples, sample_dimensions)

```python
digits = datasets.load_digits()
data_set = digits.images.reshape((len(digits.images), -1))
```

```python
example_ILS = ILS()
example_ILS.label_spreading(data_set)
```

```python
example_ILS.plot_rmin()
```

We want to segment the plot at the peaks, or segment based on differing density.

```python
suggested_indices = [190, 350, 475, 610, 715, 900, 1075, 1275, 1500]
rmin = example_ILS.rmin

plt.plot(rmin, c = "lightblue")

for ind in suggested_indices:
    plt.axvline(x = ind, c = "black")
    
plt.show()
```

```python
min_indices, labels = example_ILS.segmentation(suggested_indices)
example_ILS.label_spreading(data_set, min_indices, labels)
```

```python
example_ILS.plot_labels(use_embedding = True)
```
